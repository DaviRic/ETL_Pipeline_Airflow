```python
xls = pd.ExcelFile(path)

# Carregar a aba "Apples" ignorando a primeira linha
df_apples_clean = pd.read_excel(xls, sheet_name="Apples", skiprows=1)

# Remover linhas que contêm notas explicativas (detectadas por presença de palavras-chave)
df_apples_clean = df_apples_clean[~df_apples_clean.iloc[:, 0].astype(str).str.contains("USDA|Source|Excludes|Includes", na=False)]

# Resetar o índice após limpeza
df_apples_clean.reset_index(drop=True, inplace=True)

# Renomear colunas para facilitar a interpretação
df_apples_clean.columns = [
    "Category", "Avg_Retail_Price", "Unit", "Prep_Yield_Factor", 
    "Cup_Size", "Cup_Unit", "Avg_Price_Cup"
]

# Exibir as primeiras linhas após a limpeza
df_apples_clean.head(10)
```
